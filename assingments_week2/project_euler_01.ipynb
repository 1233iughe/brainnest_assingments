5. 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?


```python
#This function decomposes a given number in prime factors
def decompose(num):
    factors = []
    counter = 2
    while num != 1:
        #print(num, counter)
        if num % counter != 0:
            if counter % 2 == 0:
                counter += 1
            else:
                counter += 2
        else:
            factors.append(counter)
            num = num // counter

    return factors

#This function multiplies all the elements in a list
def multiplier(num_lst):
    total = 1
    for num in num_lst:
        total *= num
    return total

#This function calculates the minimum number that is evenly divisible by all the previous numbers 
def min20(num):
    #Creates a list with the decomposed numbers needed [2...N]
    num_lst = [decompose(x) for x in range(2,num + 1)]
    min_num = []
    
    #First loop takes the lists containing the factors of each number between 2 and N
    for num in num_lst:
        #Checks for every factor in the lists of prime factors if it is already registered or not
        for factor in num:
            if factor in min_num:
                #Checks if there are more ocurrences in the number(N) than in the factors list
                count_num = num.count(factor)
                count_min = min_num.count(factor)
                #print(count_num, count_min)
                if count_num > count_min:
                    for i in range(count_num - count_min):
                        min_num.append(factor)
            else:
                #Append if not registered
                min_num.append(factor)
                
    print(multiplier(min_num))
    

min20(10)
min20(20)
```

7. By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10001st prime number?


```python

def divby(num,factors):
    for factor in factors:
        if num % factor == 0:
            return False
    return True

def prime_generator(limit):
    x = [2, 3, 5, 7, 11, 13]
    i = 14
    total = 6
    while total < limit:
        if divby(i,x):
            x.append(i)
            total += 1
        i += 1
    return x


x = prime_generator(10001)
print(x[-1])
```

8. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a^2 + b^2 = c^2

For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.


```python
import math

c = 0
lst = []
#This has some issues, it produces duplicates, but works.
for i in range(1, 1000):
    for j in range(1, 1000):
        c = math.sqrt(i**2 + j**2)
        if c > 1000 or i + j > 1000:
            break
        if c.is_integer():
            lst.append([i, j, round(c)])

y = [n for n in lst if sum(n) == 1000]

for n in y:
    tot = 1
    for m in n:
        tot *= m

    print(tot)


print(y)
    
```

9. The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.


```python
#Best time of execution around 6.43 seconds
def prime_generator(lxt):
    #The seed for the list of primes is 2, could be 1 if possible_primes started at 1
    primes = [2]
    #The algorithm reduces the list lxt(possible_primes) by removing the elements divisible by the last 
    #prime number found. In this way is sure that the first element of the reduced list is the next prime number.
    #The lenght decays logarithmically(maybe?) during the proces and then converges to a rate of -1 elemnt per
    #cycle. This signals that the remaining numbers in lxt are primes, since each iteration just eliminates the current prime number 
    #itself. Also, this could be signaled by the fact that every remaining number at this point is greater than half 
    #of the limit (in this case 2,000,000 / 2 = 1,000,000) and cannot possibly have any multiples under the limit.
    #This would imply that there is at least 1 multiple of 2 remaining, and that is imposible.
    #Finally, the those remaining primes are added to lxt all at once and execution is broken.
    while lxt:
        len1 = len(lxt)
        lxt = [x for x in lxt if x % primes[-1] != 0]
        diff = len1 - len(lxt)
        #print(len(lxt), diff)
        if lxt and diff != 1:
            primes.append(lxt[0])
        else:
            primes = primes + lxt
            break
    return primes

possible_primes = [x for x in range(2,2000000)]
x = prime_generator(possible_primes)
print(sum(x))
```

10. The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?


```python
#Function generates the samllest tringle number that has a number of divisors
#equal or greater than the limit provided.
def triangle_gen(num_div):
    #This variable will contain the triangles
    triangle = 0
    #This variable will contain the nex natural number needed to create the triangle
    carrier = 1
    #Here the divisors will be stored
    divs = []
    
    
    while len(divs) <= num_div:
        #A pattern in triangular numbers is that all the jumps to more divisors occur in even numbers
        divs = [1,2]
        triangle += carrier
        carrier += 1
        
        if triangle % 2 == 0:
            #A property of division is that the divisors and quoutients are "reflected"
            #(6/2 = 3 and 6/3 = 2)
            #so its possible to save steps by storing the quotient as well as the divisor.
            
            divs.append(triangle // 2)
            for i in range(3, (triangle // 2) + 1):
                if triangle % i == 0:
                    if i >= triangle // i:
                        #Avoiding duplication of the last added divisor
                        if i not in divs:
                            divs.append(i)
                        break
                    else:
                        divs.append(i)
                        divs.append(triangle // i)
    

            
    divs.append(triangle)    
    divs.sort()   
   
    return triangle

print(triangle_gen(500))
```
